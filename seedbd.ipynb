```python
tarefas = pd.read_csv('./Tarefas.csv', sep=';')
tarefas.info()
```

```python
def converter_data_em_numero(data: str) -> int:
    partes = data.split(":")
    if len(partes) != 2:
        raise ValueError('Formato inválido. Use "HH:MM".')
    
    horas, minutos = map(int, partes)
    return horas * 60 + minutos
```

```python
tarefas['h_inicio'] = tarefas['h_inicio'].apply(converter_data_em_numero)
tarefas['h_termino'] = tarefas['h_termino'].apply(converter_data_em_numero)
```

```python
tarefas.to_sql(
  "tarefas",
  engine,
  if_exists="append",
  index=False
)
```
